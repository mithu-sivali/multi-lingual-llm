```python
import wave
import pandas as pd


root_path = "../dataset"
english_df = pd.read_csv(f'{root_path}/english/text.txt', sep="\t")
```

```python
infiles = ["english/m0005_us_m0005_00446.wav", "tamil/0.wav"]
outfile = root_path+"sounds.wav"

data= []
for infile in infiles:
    w = wave.open(f'{root_path}/{infile}', 'rb')
    data.append( [w.getparams(), w.readframes(w.getnframes())] )
    w.close()
    
output = wave.open(outfile, 'wb')
output.setparams(data[0][0])
for i in range(len(data)):
    output.writeframes(data[i][1])
output.close()
```
